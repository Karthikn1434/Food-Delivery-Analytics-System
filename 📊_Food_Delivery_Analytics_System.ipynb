**DAY1-Dataset Understanding & Basic Analysis**

```python
#CREATE DATSET
import pandas as pd

data = {
    'Order_ID':[101,102,103,104,105,106,107,108],
    'Customer':['Alex','Bob','Charlie','David','Eva','Frank','Grace','Helen'],
    'Restaurant':['Pizza Hut','Dominos','KFC','Pizza Hut','Burger King','KFC','Dominos','Pizza Hut'],
    'City':['Chennai','Mumbai','Chennai','Bangalore','Mumbai','Chennai','Bangalore','Mumbai'],
    'Amount':[500,350,400,600,450,300,550,700],
    'Delivery_Time':[30,25,35,40,20,28,32,45],
    'Rating':[5,4,4,5,3,4,5,5]
}

df = pd.DataFrame(data)

print(df)
```

```python
#Dataset Overview
print(df.head())#first 5 rows default
print("\nShape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nInfo:")
print(df.info())
```

```python
#Basic Statistics
print("Total Orders:", df['Order_ID'].count())
print("Total Revenue:", df['Amount'].sum())
print("Average Amount:", df['Amount'].mean())
print("Highest Amount:", df['Amount'].max())
print("Lowest Amount:", df['Amount'].min())
print("Average Rating:", df['Rating'].mean())
```

```python
#Filtering
print("Chennai Orders")
print(df[df['City'] == 'Chennai'])

print("\nOrders Above 500")
print(df[df['Amount'] > 500])

print("\nPizza Hut Orders")
print(df[df['Restaurant'] == 'Pizza Hut'])
```

```python
#Business Insights
print("Revenue by City")
print(df.groupby('City')['Amount'].sum())

print("\nRevenue by Restaurant")
print(df.groupby('Restaurant')['Amount'].sum())

print("\nAverage Delivery Time by City")
print(df.groupby('City')['Delivery_Time'].mean())
```

```python
#Bonus Analysis
print("Highest Rated Orders")
print(df[df['Rating'] == df['Rating'].max()])

print("\nOrders Sorted by Amount")
print(df.sort_values('Amount', ascending=False))

```

**DAY 2-Customer & Restaurant Analysis**

```python
print("Top Customers")
customer_spending = df.groupby('Customer')['Amount'].sum()
print(customer_spending.sort_values(ascending=False))

print("\nRestaurant Revenue")
restaurant_revenue = df.groupby('Restaurant')['Amount'].sum()
print(restaurant_revenue.sort_values(ascending=False))

print("\nAverage Rating by Restaurant")
print(df.groupby('Restaurant')['Rating'].mean())

print("\nRestaurant Summary")
print(df.groupby('Restaurant')['Amount'].agg(
    ['sum', 'mean', 'max', 'min']
))

print("\nCity Analysis")
print(df.groupby('City')['Amount'].agg(
    ['sum', 'mean', 'count']
))

print("\n5-Star Orders")
print(df[df['Rating'] == 5])

print("\nOrders Sorted by Amount")
print(df.sort_values('Amount', ascending=False))

print("\nLeast Spending Customer")
print(customer_spending.sort_values())
```

**DAY 3**- **Data Cleaning & EDA**


```python

print("Missing Values")
print(df.isnull().sum())

print("\nDuplicate Rows")
print(df.duplicated().sum())

print("\nData Types")
print(df.dtypes)

print("\nUnique Cities")
print(df['City'].unique())

print("\nOrders by City")
print(df['City'].value_counts())

print("\nOrders by Restaurant")
print(df['Restaurant'].value_counts())

print("\nSummary Statistics")
print(df.describe())

print("\nNumber of Unique Restaurants")
print(df['Restaurant'].nunique())
```

**Day 4: Advanced EDA & Feature Engineering**

```python
#creates a new column from existing data.This is called Feature Engineering.

df['Order_Type'] = df['Amount'].apply(
    lambda x: 'High Value' if x > 500 else 'Regular'
)

print("Order Type")
print(df[['Customer','Amount','Order_Type']])

print("\nOrder Type Count")
print(df['Order_Type'].value_counts())

print("\nAverage Rating by City")
print(df.groupby('City')['Rating'].mean())

print("\nAverage Delivery Time by City")
print(df.groupby('City')['Delivery_Time'].mean())

print("\nAverage Amount by Restaurant")
print(df.groupby('Restaurant')['Amount'].mean())

print("\nHigh Value Orders")
print(df[df['Amount'] > 500])

print("\nTop Rated Restaurant")
print(df.groupby('Restaurant')['Rating'].mean().sort_values(ascending=False))

print("\nFastest Delivery City")
print(df.groupby('City')['Delivery_Time'].mean().sort_values())
```

**Day 5: Matplotlib Visualization.**

```python
import matplotlib.pyplot as plt

# Revenue by Restaurant
restaurant_revenue = df.groupby('Restaurant')['Amount'].sum()
print("Revenue by Restaurant")
print(restaurant_revenue)
restaurant_revenue.plot(kind='bar')
plt.title('Revenue by Restaurant')
plt.xlabel('Restaurant')
plt.ylabel('Revenue')
plt.show()

# Revenue by City
city_revenue = df.groupby('City')['Amount'].sum()
print("\nRevenue by City")
print(city_revenue)
city_revenue.plot(kind='bar')
plt.title('Revenue by City')
plt.xlabel('City')
plt.ylabel('Revenue')
plt.show()

# Orders by Restaurant
restaurant_orders = df['Restaurant'].value_counts()
print("\nOrders by Restaurant")
print(restaurant_orders)
restaurant_orders.plot(kind='bar')
plt.title('Orders by Restaurant')
plt.xlabel('Restaurant')
plt.ylabel('Orders')
plt.show()

# Rating Distribution
rating_dist = df['Rating'].value_counts()
print("\nRating Distribution")
print(rating_dist)
rating_dist.plot(kind='pie', autopct='%1.1f%%')
plt.title('Rating Distribution')
plt.ylabel('')
plt.show()

# Customer Spending
customer_spending = df.groupby('Customer')['Amount'].sum()
print("\nCustomer Spending")
print(customer_spending)
customer_spending.sort_values(ascending=False).plot(kind='bar')
plt.title('Customer Spending')
plt.xlabel('Customer')
plt.ylabel('Amount')
plt.show()

# Business Insights
print("\n===== BUSINESS INSIGHTS =====")

print("Highest Revenue Restaurant:",
      restaurant_revenue.idxmax(),
      "-", restaurant_revenue.max())

print("Highest Revenue City:",
      city_revenue.idxmax(),
      "-", city_revenue.max())

print("Most Orders Restaurant:",
      restaurant_orders.idxmax(),
      "-", restaurant_orders.max())

five_star = (df['Rating'] == 5).mean() * 100
print("5-Star Rating Percentage:",
      round(five_star, 2), "%")

print("Highest Spending Customer:",
      customer_spending.idxmax(),
      "-", customer_spending.max())
```

**Day 6: NumPy + SQL Analysis**

```python
import numpy as np
import sqlite3
import pandas as pd

print("===== NUMPY ANALYSIS =====")

amounts = np.array(df['Amount'])
ratings = np.array(df['Rating'])
delivery = np.array(df['Delivery_Time'])

print("Shape:", amounts.shape)
print("Datatype:", amounts.dtype)
print("Size:", amounts.size)

print("Total Revenue:", np.sum(amounts))
print("Average Revenue:", np.mean(amounts))
print("Highest Order:", np.max(amounts))
print("Lowest Order:", np.min(amounts))

print("Average Rating:", np.mean(ratings))
print("Highest Rating:", np.max(ratings))
print("Lowest Rating:", np.min(ratings))

print("Average Delivery Time:", np.mean(delivery))
print("Fastest Delivery:", np.min(delivery))
print("Slowest Delivery:", np.max(delivery))

print("\n===== SQL ANALYSIS =====")

conn = sqlite3.connect(':memory:')
#craete orders table
df.to_sql('orders', conn, index=False, if_exists='replace')

print("\n1. All Orders")
print(pd.read_sql("""
SELECT *
FROM orders
""", conn))

print("\n2. Orders with Rating 5")
print(pd.read_sql("""
SELECT *
FROM orders
WHERE Rating = 5
""", conn))

print("\n3. Orders Above ₹500")
print(pd.read_sql("""
SELECT Customer, Amount
FROM orders
WHERE Amount > 500
""", conn))

print("\n4. Revenue by Restaurant")
print(pd.read_sql("""
SELECT Restaurant,
SUM(Amount) AS Revenue
FROM orders
GROUP BY Restaurant
""", conn))

print("\n5. Revenue by City")
print(pd.read_sql("""
SELECT City,
SUM(Amount) AS Revenue
FROM orders
GROUP BY City
""", conn))

print("\n6. Average Rating by Restaurant")
print(pd.read_sql("""
SELECT Restaurant,
AVG(Rating) AS Avg_Rating
FROM orders
GROUP BY Restaurant
""", conn))

print("\n7. Top 3 Highest Orders")
print(pd.read_sql("""
SELECT Customer, Amount
FROM orders
ORDER BY Amount DESC
LIMIT 3
""", conn))

print("\n8. Lowest to Highest Orders")
print(pd.read_sql("""
SELECT Customer, Amount
FROM orders
ORDER BY Amount ASC
""", conn))
```
